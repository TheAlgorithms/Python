```python
df_data = df[['CONTENT', 'CLASS']]
df_data
```

```python
x = df_data.CONTENT
y = df_data.CLASS
```

```python
cv = CountVectorizer() #The CountVectorizer provides a simple way to both tokenize a collection of text documents and build a vocabulary of known words, but also to encode new documents using that vocabulary. 
ex = cv.fit_transform(["Great song but check this out","What is this song?"])
```

```python
ex.toarray()
```

```python
cv.get_feature_names()
```

```python
cv = CountVectorizer()
corpus = x
X = cv.fit_transform(corpus)
```

```python
X.toarray()
```

```python
X.shape
```

```python
X.size
```

```python
cv.get_feature_names()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
print(X_train.size)
print(X_test.size)
print(y_train.size)
print(y_test.size)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train
```

```python
#IMPLEMENTING NAIVE-BAYES CLASSIFIER
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(X_train, y_train)
```

```python
print(clf.intercept_)
print(clf.coef_)
print(clf.classes_)
print(clf.score(X_train, y_train))
```

```python
print(clf.score(X_test, y_test))
```

```python
print("Accuracy of Model",clf.score(X_test,y_test)*100,"%")
```

```python
pred = clf.predict(X_test)
pred
```
