```python
price = 'Rs.35000 only'
print(int(price[3:-5]))
print(int(price[3:3+5]))
```

```python
word = "William Shakespeare"
```

slicing syntax

`var[start:stop]`

`var[start:stop:step]`

```python
word[:7:2] # start to 6 index, jump 2 index at a time
```

```python
word[::2] # start to end, jump 2 index at a time
```

```python
# reverse a string
word[::-1]
```
